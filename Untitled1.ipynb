```python
from keras.models import load_model
import numpy as np
from sklearn.metrics import precision_recall_curve, f1_score, auc
import matplotlib.pyplot as plt

```

```python
def mean_squared_loss(x1, x2):
    diff = x1 - x2
    a, b, c, d, e = diff.shape
    n_samples = a * b * c * d * e
    sq_diff = diff ** 2
    Sum = sq_diff.sum()
    dist = np.sqrt(Sum)
    mean_dist = dist / n_samples
    return mean_dist

```

```python
# Load the model
model = load_model('AnomalyDetector.h5')

# Load the testing data
X_test = np.load('tester.npy')
frames = X_test.shape[2]
frames = frames - frames % 10
X_test = X_test[:, :, :frames]
X_test = X_test.reshape(-1, 227, 227, 10)
X_test = np.expand_dims(X_test, axis=4)

```

```python
# Initialize variables
best_threshold = None
best_f1_score = 0.0
ground_truth_labels = []  # Store the ground truth labels
anomaly_scores = []  # Store the anomaly scores

# thresholds = [0.00024,0.000243,0.000245,0.000249]
thresholds = [0.00025, 0.000253,0.000255, 0.000259, 0.000260, 0.000265]

# 1. Low Threshold (More Positives): If you set a low threshold, more instances are predicted as positive (anomalies). 
#     This can increase recall (finding more actual anomalies), but it might also lower precision (more false positives). 
#     Consequently, the F1-score might decrease as well since it considers both precision and recall.

# 2. High Threshold (More Negatives): On the other hand, if you set a high threshold, fewer instances are predicted as positive. 
#     This can increase precision (more confident predictions are correct), but it might decrease recall (missing some actual anomalies). 
#     Again, the F1-score could decrease as a result.

# 3. Optimal Threshold (Balanced F1-Score): The optimal threshold is where you achieve the best trade-off between precision and recall. 
#     This is the point where the F1-score is maximized. 
#     In other words, it's the threshold that yields the best balance between accurately identifying anomalies and minimizing false alarms.


```

```python
plot_max_f1_score = []
# Iterate through different thresholds and calculate anomaly scores
for threshold in thresholds:
    counter = 0
    anomaly_count = 0
    for number, bunch in enumerate(X_test):
        n_bunch = np.expand_dims(bunch, axis=0)
        reconstructed_bunch = model.predict(n_bunch)
        loss = mean_squared_loss(n_bunch, reconstructed_bunch)

        if loss > threshold:
            anomaly_count += 1
            ground_truth_labels.append(1)  # Anomalous
            anomaly_scores.append(loss)  # Store the anomaly score
        else:
            ground_truth_labels.append(0)  # Normal
            anomaly_scores.append(loss)  # Store the anomaly score

    # Calculate precision, recall, and F1-score for each threshold
    precision, recall, _ = precision_recall_curve(ground_truth_labels, anomaly_scores)
    average_precision = auc(recall, precision)
    f1_scores = 2 * (precision * recall) / (precision + recall)
    max_f1_score = np.max(f1_scores)
    plot_max_f1_score.append(max_f1_score) 

    # Compare F1-score to find the best threshold
    if max_f1_score > best_f1_score:
        best_f1_score = max_f1_score
        best_threshold = threshold

    print("Threshold: {:.6f}, Max F1-score: {:.6f}".format(threshold, max_f1_score))

print("Best Threshold: {:.6f}, Best F1-score: {:.6f}".format(best_threshold, best_f1_score))



# Set the best threshold for anomaly detection
threshold = best_threshold

# # Initialize variables for accuracy calculation
correct_predictions = 0
total_samples = len(ground_truth_labels)

# Calculate accuracy by comparing anomaly scores with the threshold
for i in range(total_samples):
    if anomaly_scores[i] > threshold and ground_truth_labels[i] == 1:
        correct_predictions += 1
    elif anomaly_scores[i] <= threshold and ground_truth_labels[i] == 0:
        correct_predictions += 1

# Calculate accuracy
accuracy = correct_predictions / total_samples
print("Accuracy: {:.2f}%".format(accuracy * 100))
print("Precision: {:.4f}".format(precision[np.argmax(f1_scores)]))
print("Recall: {:.4f}".format(recall[np.argmax(f1_scores)]))
print("Best F1-score: {:.4f}".format(best_f1_score))
print("AUC: {:.4f}".format(average_precision))

# Detect anomalies using the selected threshold
flag = 0  # Overall video flag
for number, bunch in enumerate(X_test):
    n_bunch = np.expand_dims(bunch, axis=0)
    reconstructed_bunch = model.predict(n_bunch)
    loss = mean_squared_loss(n_bunch, reconstructed_bunch)

    if loss > threshold:
        print("Anomalous bunch of frames at bunch number {}".format(number))
        counter = counter + 1
        print("Bunch number: ", counter)
        flag = 1
    else:
        print('No anomaly')
        counter = counter + 1
        print("Bunch number: ", counter)

if flag == 1:
    print("Anomalous Events detected")
else:
    print("No anomaly detected")

print("\nCorrupted and unreadable bunches were ignored")
# Plot the Precision-Recall Curve
plt.figure(figsize=(10, 6))
plt.plot(recall, precision, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.grid()
plt.show()

# Plot the F1-Score Curve against different thresholds
plt.figure(figsize=(10, 6))
plt.plot(thresholds, plot_max_f1_score, marker='.')
plt.xlabel('Threshold')
plt.ylabel('F1-Score')
plt.title('F1-Score Curve')
plt.grid()
plt.show()
```
